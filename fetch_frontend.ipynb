```python
import json
import urllib.request

base = 'https://api.github.com/repos/Luisceron0/ElevaForge'

# Get tree
tree_url = f'{base}/git/trees/main?recursive=1'
req = urllib.request.Request(tree_url, headers={'User-Agent': 'Python'})
with urllib.request.urlopen(req) as resp:
    tree_data = json.loads(resp.read().decode())

files = [f['path'] for f in tree_data['tree'] if f['type'] == 'blob']
print('=== FILE TREE ===')
print(json.dumps(files, indent=2))

# Filter key files
key_files = [f for f in files if (
    f == 'package.json' or
    f.endswith('.env.example') or
    f.endswith('.env.local.example') or
    'api' in f or
    'lead' in f.lower() or
    'contact' in f.lower() or
    'form' in f.lower() or
    'supabase' in f or
    'types' in f or
    'lib/' in f or
    'hooks/' in f or
    'utils/' in f or
    'services/' in f or
    'actions/' in f
)]

print('\n=== KEY FILES ===')
print(json.dumps(key_files, indent=2))

for file in key_files:
    url = f'https://raw.githubusercontent.com/Luisceron0/ElevaForge/main/{file}'
    req2 = urllib.request.Request(url, headers={'User-Agent': 'Python'})
    try:
        with urllib.request.urlopen(req2) as resp2:
            content = resp2.read().decode()
            print(f'\n=== FILE: {file} ===')
            print(content)
            print(f'=== END: {file} ===\n')
    except Exception as e:
        print(f'Failed to fetch {file}: {e}')
```
